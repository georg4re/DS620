```python
#import the needed libraries
import networkx as nx
```

### Create the Graph Object 

```python
# Create a networkx graph object
my_graph = nx.Graph() 
```

### Add the data to the graph object

```python
# Add edges to to the graph object
# Each tuple represents an edge between two nodes
my_graph.add_edges_from([
                        ("Jane","Ike"), 
                        ("Ike","Heather"), 
                        ("Heather","Fernando"), 
                        ("Heather", "Garth"), 
                        ("Fernando", "Garth"),
                        ("Fernando", "Carol"),
                         ("Fernando", "Diane"),
                         ("Fernando", "Andre"),
                         ("Garth", "Diane"),
                         ("Garth", "Ed"),
                         ("Garth", "Beverly"),
                         ("Diane","Carol"),
                         ("Diane", "Ed"),
                         ("Diane","Beverly"),
                         ("Diane","Andre"),
                          ("Carol", "Andre"),
                         ("Andre", "Beverly"),
                         ("Beverly", "Ed")
                        ])
```

### Draw the graph

```python
# Draw the resulting graph
nx.draw(my_graph, with_labels=True, font_weight='bold')
```
